```python
data = data.drop_duplicates()
data.duplicated().sum()
```

```python
data = data.drop(columns = "Not_Useful_Column")
data.info()
```

```python
data[['First_Name','Last_Name']]
```

```python
data['Last_Name'] = data['Last_Name'].str.strip()
data['Last_Name'] = data['Last_Name'].str.strip("...")
data['Last_Name'] = data['Last_Name'].str.strip("/")
data['Last_Name'] = data['Last_Name'].str.strip("_")
```

```python
data[['First_Name','Last_Name']]
```

```python
data['Phone_Number']
```

```python
data['Phone_Number'] = data['Phone_Number'].replace('[^a-zA-Z0-9]','')
data['Phone_Number'] = data['Phone_Number'].apply(lambda x : str(x))
data['Phone_Number'] = data['Phone_Number'].apply(lambda x : x[0:3]+'-'+x[4:7]+'-'+x[8:10])
data['Phone_Number'] = data['Phone_Number'].str.replace('nan--',' ')
data['Phone_Number'] = data['Phone_Number'].str.replace('N/a--',' ')
data['Phone_Number']
```

```python
data[['Street_Address','State','Zip_Code']] = data['Address'].str.split(',',expand=True)
data.head()
```

```python
data['Paying Customer'] = data['Paying Customer'].replace({'Yes':'Y','No':'N'})
data['Do_Not_Contact'] = data['Do_Not_Contact'].replace({'Yes':'Y','No':'N'})
data = data.fillna(' ')
data
```

```python
data.to_csv('customer_call_list.csv')
```
